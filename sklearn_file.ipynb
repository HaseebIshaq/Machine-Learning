```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif
from scipy.stats import ttest_ind
from sklearn.preprocessing import StandardScaler
```

# avg_rating is the target column

```python
# 1. Import dataset
df = pd.read_csv('tripadvisor_european_restaurants.csv')
df
```

```python
# 2. Preprocessing
# Example: Standardize data
scaler = StandardScaler()
numerical_columns = ['latitude', 'longitude', 'open_days_per_week', 'open_hours_per_week', 'working_shifts_per_week',
                     'avg_rating', 'total_reviews_count', 'reviews_count_in_default_language',
                     'excellent', 'very_good', 'average', 'poor', 'terrible', 'food', 'service', 'value', 'atmosphere']

df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
# 3. Handle missing values manually
# For numerical columns, let's fill missing values with the mean of each column.
numerical_missing_columns = ['latitude', 'longitude', 'open_days_per_week', 'open_hours_per_week', 'working_shifts_per_week',
                             'avg_rating', 'total_reviews_count', 'reviews_count_in_default_language',
                             'excellent', 'very_good', 'average', 'poor', 'terrible', 'food', 'service', 'value', 'atmosphere']

for column in numerical_missing_columns:
    df[column].fillna(df[column].mean(), inplace=True)

# For categorical columns, let's fill missing values with the most frequent value (mode) of each column.
categorical_missing_columns = ['claimed', 'awards', 'popularity_detailed', 'popularity_generic', 'top_tags', 'price_level',
                               'price_range', 'meals', 'cuisines', 'special_diets', 'features', 'default_language']

for column in categorical_missing_columns:
    mode_value = df[column].mode()[0]  # Get the mode value
    df[column].fillna(mode_value, inplace=True)
```

```python
# Convert all string values to float by use of label encoder
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Specify the list of columns that you want to encode
categorical_columns = ['restaurant_link', 'restaurant_name', 'original_location', 'country', 'region', 'province', 'city', 'address',
                       'claimed', 'awards', 'popularity_detailed', 'popularity_generic', 'top_tags', 'price_level',
                       'price_range', 'meals', 'cuisines', 'special_diets', 'features', 'default_language', 'vegetarian_friendly',
                       'vegan_options', 'gluten_free', 'original_open_hours', 'keywords']

for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
# 4. Apply regression-oriented feature selection techniques
# ANOVA F-value for regression
from sklearn.feature_selection import SelectKBest, f_regression, mutual_info_regression

k = 2
selector_f = SelectKBest(f_regression, k=k)
X_new_f = selector_f.fit_transform(df[numerical_columns], df['avg_rating'])
selected_features_f = df[numerical_columns].columns[selector_f.get_support()]

# Mutual Information for regression
selector_m = SelectKBest(mutual_info_regression, k=k)
X_new_m = selector_m.fit_transform(df[numerical_columns], df['avg_rating'])
selected_features_m = df[numerical_columns].columns[selector_m.get_support()]

print(f"Features selected by ANOVA F-value (Regression): {selected_features_f}")
print(f"Features selected by Mutual Information (Regression): {selected_features_m}")
```

```python
# 5. Correlation analysis
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
# 5. Visualization
import seaborn as sns

plt.figure(figsize=(12, 10))

# Create a heatmap of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Numerical columns vs. avg_rating (Scatter Plots)
numerical_columns = ['latitude', 'longitude', 'open_days_per_week', 'open_hours_per_week', 'working_shifts_per_week',
                     'total_reviews_count', 'reviews_count_in_default_language',
                     'excellent', 'very_good', 'average', 'poor', 'terrible', 'food', 'service', 'value', 'atmosphere']

for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    plt.scatter(df[column], df['avg_rating'], alpha=0.5)
    plt.xlabel(column)
    plt.ylabel('Avg Rating')
    plt.title(f'{column} vs Avg Rating')
    plt.grid(True)
    plt.savefig(f"{column} vs Avg Rating.pdf", dpi=300, format="pdf")
    plt.show()
```
